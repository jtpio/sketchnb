# p5.js Sketch

A sample notebook using the p5.js kernel.

```p5js
function setup() {
  createCanvas(innerWidth, innerHeight);
}

function draw() {
  background(220);

  // Follow the mouse
  fill(63, 81, 181);
  noStroke();
  circle(mouseX, mouseY, 50);

  // Trail effect
  fill(233, 30, 99, 150);
  circle(mouseX + 20, mouseY + 20, 30);
  circle(mouseX - 20, mouseY - 20, 30);
}
```
