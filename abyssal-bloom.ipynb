# Abyssal Bloom

In the deepest reaches of the ocean, beyond the last fading photon of sunlight, life invents its own luminance. Bioluminescent organisms — jellyfish trailing veils of cold fire, siphonophores pulsing in coordinated chains, dinoflagellates igniting in shimmering cascades — trace invisible currents in ribbons of light. Their movements are governed entirely by forces they cannot perceive. They do not choose their paths. They surrender to the flow, and in that surrender, create patterns of startling beauty.

This algorithm mirrors that process. Fifteen hundred particles are released into a time-evolving vector field derived from three-dimensional Perlin noise — two spatial dimensions and one temporal. The field shifts continuously, its currents drifting and reforming like tidal flows over a reef. Each particle follows the field blindly, drawing a thin luminous filament behind it that fades over time, leaving ghostly trails that reveal the field's hidden structure before dissolving back into the dark.

The color palette draws from the deep ocean's own spectrum — the teals of bioluminescent bacteria, the electric blues of comb jellies refracting their own cilia-light, the violet phosphorescence of deep-sea cephalopods signaling across the dark. Hues shift with each particle's direction of travel, so that convergent flows bloom in shared color while divergent streams reveal the full chromatic range of the abyss.

The seed is the ocean. Change it, and the currents shift — new vortices appear, old corridors collapse — but the character remains. Every variation is recognizably a bloom from the same abyss, a living system that never repeats and never rests.

```p5js
let seed = 42;
let particles = [];

let params = {
  count: 1500,
  noiseScale: 0.003,
  timeScale: 0.0008,
  fieldTwist: 4,
  strokeAlpha: 30,
  weight: 0.8,
  hueStart: 170,
  hueEnd: 310,
  trailAlpha: 3,
  speedMin: 1,
  speedMax: 3,
};
```

```p5js
class Particle {
  constructor() {
    this.pos = createVector(random(width), random(height));
    this.prev = this.pos.copy();
    this.speed = random(params.speedMin, params.speedMax);
    this.baseHue = random(params.hueStart, params.hueEnd);
  }

  update() {
    this.prev.set(this.pos);

    let angle = noise(
      this.pos.x * params.noiseScale,
      this.pos.y * params.noiseScale,
      frameCount * params.timeScale
    ) * TWO_PI * params.fieldTwist;

    this.pos.x += cos(angle) * this.speed;
    this.pos.y += sin(angle) * this.speed;

    if (this.pos.x < 0 || this.pos.x > width ||
        this.pos.y < 0 || this.pos.y > height) {
      this.pos.set(random(width), random(height));
      this.prev.set(this.pos);
    }
  }

  display() {
    let dx = this.pos.x - this.prev.x;
    let dy = this.pos.y - this.prev.y;
    let angle = atan2(dy, dx);
    let hue = (this.baseHue + angle * 8 + 360) % 360;

    stroke(hue, 65, 85, params.strokeAlpha);
    strokeWeight(params.weight);
    line(this.prev.x, this.prev.y, this.pos.x, this.pos.y);
  }
}
```

```p5js
function setup() {
  createCanvas(innerWidth, innerHeight);
  randomSeed(seed);
  noiseSeed(seed);
  colorMode(HSB, 360, 100, 100, 100);
  background(230, 50, 6);

  for (let i = 0; i < params.count; i++) {
    particles.push(new Particle());
  }
}

function draw() {
  noStroke();
  fill(230, 50, 6, params.trailAlpha);
  rect(0, 0, width, height);

  for (let p of particles) {
    p.update();
    p.display();
  }
}
```
